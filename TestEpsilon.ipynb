```python
import numpy as np
from material_function import moebius_eps_func
```

```python
x = np.linspace(-10, 10, 101)
y = np.linspace(-10, 10, 101)
z = np.linspace(-3, 3, 101)
result = np.zeros((x.shape[0], y.shape[0], z.shape[0]))
for i in range(x.shape[0]):
    for j in range(y.shape[0]):
        for k in range(z.shape[0]):
            result[i, j, k] = moebius_eps_func(1, 0, 3, 2, 4, np.pi, x[i], y[j], z[k])
```

```python
from matplotlib import pyplot as plt

plt.pcolormesh(y, z, result[len(x) // 2, :, :].T)
plt.xlabel("y")
plt.ylabel("z")
plt.colorbar()
plt.gca().set_aspect('equal')
```

```python
plt.pcolormesh(x, y, result[:, :, len(z) // 2].T)
plt.xlabel("x")
plt.ylabel("y")
plt.colorbar()
plt.gca().set_aspect('equal')
```

```python
plt.pcolormesh(y, z, result[:, len(y) // 2, :].T)
plt.xlabel("x")
plt.ylabel("z")
plt.colorbar()
plt.gca().set_aspect('equal')
```
